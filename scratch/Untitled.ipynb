```python
import audata
import h5py as h5
from os import listdir
import pandas as pd
import time
from datetime import datetime
```

```python
files = listdir('/zfsauton/data/public/gwelter/AUView/projects/conditionc/originals')
```

```python
pd.set_option('display.float_format', lambda x: '%.6f' % x)
```

```python
df = pd.DataFrame(columns = ['filename', 'oldval', 'newval', 'oldtime', 'newtime', 'diff', 'addtoann'])
for i in range(len(files)):
    f = files[i]
    try:
        oldf = h5.File('/zfsauton/data/public/gwelter/AUView/projects/old/ConditionC/originals/'+f, 'r')
        newf = audata.File.open('/zfsauton/data/public/gwelter/AUView/projects/conditionc/originals/'+f)
    except:
        continue
    if 'numerics' in oldf.keys():
        
        oldentry = oldf['numerics/HR.HR/data'][0]
        newentry = newf['data/numerics/HR.HR'][0]
        
        oldts = oldentry[0] + oldf['meta'].attrs['baseTime']
        newts = newentry['time'][0].timestamp()
        diff = (newts-oldts)/60/60
        
        addtoann = oldf['meta'].attrs['baseTime'] + (newts-oldts)
        
        df.loc[i] = [f, oldentry[1], newentry['value'][0], datetime.fromtimestamp(oldts), newentry['time'][0], diff, addtoann]

df
```

```python
df = pd.DataFrame(columns = ['filename', 'oldts', 'newts', 'diff'])
for i in range(len(files)):
    f = files[i]
    try:
        oldf = h5.File('/zfsauton/data/public/gwelter/AUView/projects/old/ConditionC/originals/'+f, 'r')
        newf = audata.File.open('/zfsauton/data/public/gwelter/AUView/projects/conditionc/originals/'+f)
    except:
        continue
    oldts = oldf['meta'].attrs['baseTime']
    newts = newf.time_reference.timestamp()
    df.loc[i] = [f, datetime.fromtimestamp(oldts), newf.time_reference, round((newts-oldts)/60/60, 2)]
df.sort_values(by=['diff'])
```
