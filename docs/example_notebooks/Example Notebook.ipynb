```python
# List available projects
auviewer.api.listAvailableProjects()
```

```python
# Load a project
p = auviewer.api.loadProject(1)
```

```python
# Check what methods are available for the project
help(p)
```

```python
# List pattern sets associated with the project
p.listPatternSets()
```

```python
# Get a pattern set which contains annotations
ps = p.getPatternSet(10)
```

```python
# Check what methods are available for the pattern set
help(ps)
```

```python
# Retrieve the patterns associated with the pattern set
patterns = ps.getPatterns()
display(patterns.head())
```

```python
# Retrieve the annotations from the pattern set
annotations = ps.getAnnotations()
display(annotations.head())
```
