```python
# Get project id=1
p = getProjectByID(2)
print(p)
```

```python
# Create an pattern set
patternset = p.createPatternSet('Initial CRI Real v Art Annotation')

# Run pattern detection
hrPatterns = p.detectPatterns(
    type='patterndetection',
    series='/data/numerics/HR.HR:value',
    thresholdlow=58,
    thresholdhigh=110,
    duration=10,
    persistence=.7,
    maxgap=1800
)
rrPatterns = p.detectPatterns(
    type='patterndetection',
    series='/data/numerics/RR.RR:value',
    thresholdlow=10,
    thresholdhigh=29,
    duration=300,
    persistence=.7,
    maxgap=300
)
spo2Patterns = p.detectPatterns(
    type='patterndetection',
    series='/data/numerics/SpO₂T.SpO₂T:value',
    thresholdlow=90,
    thresholdhigh=None,
    duration=300,
    persistence=.7,
    maxgap=300
)

# Add the patterns to the pattern set we created
patternset.addPatterns(hrPatterns)
patternset.addPatterns(rrPatterns)
patternset.addPatterns(spo2Patterns)
```
